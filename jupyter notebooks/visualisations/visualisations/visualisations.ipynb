### Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
df_08.shape, df_18.shape
```

Compare the distributions of greenhouse gas score in 2008 and 2018.

```python
df_08['greenhouse_gas_score'].hist(figsize=(8,8))
```

```python
df_18['greenhouse_gas_score'].hist(figsize=(8,8))
```

How has the distribution of combined mpg changed from 2008 to 2018?

```python
df_08['cmb_mpg'].hist(figsize=(8,8))

```

```python
df_18['cmb_mpg'].hist(figsize=(8,8))
```

Describe the correlation between displacement and combined mpg.
[Add a Trendline in plt scatter graph](https://stackoverflow.com/questions/41635448/how-can-i-draw-scatter-trend-line-on-matplot-python-pandas/41635626)

```python
y=df_08['displ']
x=df_08['cmb_mpg']

plt.scatter(x, y)
plt.xlabel('cmb_mpg')
plt.ylabel('dipl')


z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
```

```python
y=df_18['displ']
x=df_18['cmb_mpg']

plt.scatter(x, y)
plt.xlabel('cmb_mpg')
plt.ylabel('dipl')


z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()

```
